```python
# Cell 2: Import libraries
# !pip install -U transformers datasets evaluate accelerate librosa jiwer soundfile

import os
import torch
# import torchaudio
import json
from datasets import load_dataset, Dataset
from transformers import WhisperProcessor, WhisperForConditionalGeneration, TrainingArguments, Trainer, Seq2SeqTrainer
from transformers.trainer_callback import TrainerCallback
import evaluate
from transformers import TrainingArguments
import numpy as np
from datasets import Dataset, load_dataset
import json

# Function to load JSONL into Python list
# def load_jsonl(path):
#     with open(path, "r", encoding="utf-8") as f:
#         return [json.loads(line) for line in f]

# train_path = "/home/sagemaker-user/train.jsonl"
# valid_path = "/home/sagemaker-user/valid.jsonl"
# Convert to Hugging Face Dataset
from datasets import load_dataset
# !pip install soundfile

train_dataset = load_dataset("json", data_files="./train.jsonl")["train"]
valid_dataset = load_dataset("json", data_files="./valid.jsonl")["train"]
```

```python
# unzip the data.zip file and put it in a 'data' folder, and rename audio files
import zipfile
import os

zip_file_path = 'data.zip' # Ensure this path is correct
extract_dir = 'data'

# Create the data directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir) # Extracts to the 'data' directory

print(f"✅ Extracted data.zip to {extract_dir}")

# Rename audio files by removing leading zeros
for filename in os.listdir(extract_dir):
    if filename.endswith(".wav"):
        try:
            # Assuming the format is like '01.wav', '02.wav', etc.
            # This might need adjustment based on actual filenames
            name, ext = os.path.splitext(filename)
            new_name = str(int(name)) + ext
            old_path = os.path.join(extract_dir, filename)
            new_path = os.path.join(extract_dir, new_name)
            if old_path != new_path: # Avoid renaming if already in the correct format
                os.rename(old_path, new_path)
                # print(f"Renamed {filename} to {new_name}") # Uncomment for debugging
        except ValueError:
            # Handle cases where filename is not a number (e.g., 'SENT_0905.wav')
            # print(f"Skipping renaming for {filename}") # Uncomment for debugging
            pass

print("✅ Renamed audio files in data/ (removed leading zeros if applicable)")
```

```python
from transformers import WhisperForConditionalGeneration, WhisperProcessor
import torchaudio

processor = WhisperProcessor.from_pretrained("openai/whisper-large-v3-turbo")
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-large-v3-turbo")
def preprocess(example):
    waveform, sr = torchaudio.load(example["audio_filepath"])
    # Resample if necessary
    if sr != 16000:
        resampler = torchaudio.transforms.Resample(orig_freq=sr, new_freq=16000)
        waveform = resampler(waveform)
        sr = 16000

    inputs = processor.feature_extractor(waveform[0], sampling_rate=sr, return_tensors="pt")
    input_features = inputs.input_features[0]

    labels = processor.tokenizer(example["human_transcript"]).input_ids
    example["input_features"] = input_features
    example["labels"] = labels
    return example

train_dataset = train_dataset.map(preprocess)
valid_dataset = valid_dataset.map(preprocess)
```

```python
def data_collator(features):
    # Ensure input_features are also converted to tensors
    input_features = torch.stack([torch.tensor(f["input_features"]) for f in features])
    label_ids = [f["labels"] for f in features]
    labels = processor.tokenizer.pad({"input_ids": label_ids}, return_tensors="pt").input_ids
    labels[labels == processor.tokenizer.pad_token_id] = -100
    return {"input_features": input_features, "labels": labels}
    wer_metric = evaluate.load("wer")
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    pred_ids = np.argmax(logits, axis=-1)
    pred_str = processor.tokenizer.batch_decode(pred_ids, skip_special_tokens=True)
    label_str = processor.tokenizer.batch_decode(labels, skip_special_tokens=True)
    return {"wer": wer_metric.compute(predictions=pred_str, references=label_str)}
# Cell 8: Trainer callback for logging
class LoggingCallback(TrainerCallback):
    def on_log(self, args, state, control, logs=None, **kwargs):
        if logs:
            print(f"Epoch {state.epoch:.2f} - Logs: {logs}")
```

```python
from transformers import Seq2SeqTrainingArguments
training_args = Seq2SeqTrainingArguments(
    output_dir="./fine_tuned_whisper",
    per_device_train_batch_size=1,
    num_train_epochs=1,
    learning_rate=5e-5,
    logging_dir="./logs",
    predict_with_generate=True  # this is important for WER calculation
)
# Now initialize trainer
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=valid_dataset,
    tokenizer=processor,  # processor includes feature extractor + tokenizer
    data_collator=data_collator,
    compute_metrics=compute_metrics
)
```

```python
# !pip install bitsandbytes
# !pip install peft
from peft import get_peft_model, LoraConfig, TaskType

# LoRA Config
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["out_proj", "k_proj"],  # change to ["out_proj", "k_proj"] if CPU
    lora_dropout=0.1,
    bias="none",
    task_type=TaskType.SEQ_2_SEQ_LM
)

# Apply LoRA to model
model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
```

```python
trainer.train()
# import torch
# torch.cuda.empty_cache()
# c1cec1abc4a9c3e91f48ff71fb39c80d898d4833
# Clear cache
# torch.cuda.empty_cache()

# # Optional: reset memory stats
# torch.cuda.reset_peak_memory_stats()
```

```python
import zipfile
import os
from google.colab import files

# Define the directories to save the model and processor
adapter_dir = "turbo_adapter"
processor_dir = "turbo_whisper"

# Save the model adapter and processor
print(f"Saving model adapter to {adapter_dir}...")
model.save_pretrained(adapter_dir)
print(f"✅ Model adapter saved to {adapter_dir}")

print(f"Saving processor to {processor_dir}...")
processor.save_pretrained(processor_dir)
print(f"✅ Processor saved to {processor_dir}")


# Define the zip filenames
adapter_zip_filename = f"{adapter_dir}.zip"
processor_zip_filename = f"{processor_dir}.zip"

# Function to zip a directory
def zip_directory(directory_path, zip_path):
    print(f"Compressing '{directory_path}' into '{zip_path}'...")
    with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        if os.path.exists(directory_path):
            for root, _, files in os.walk(directory_path):
                for file in files:
                    file_path = os.path.join(root, file)
                    arcname = os.path.relpath(file_path, '.') # Keep directory structure
                    zipf.write(file_path, arcname)
        else:
            print(f"Warning: Directory '{directory_path}' not found. Skipping.")
    print(f"✅ Created '{zip_path}'")

# Zip the directories separately
zip_directory(adapter_dir, adapter_zip_filename)
zip_directory(processor_dir, processor_zip_filename)


# Provide download links
print(f"⬇️ Downloading {adapter_zip_filename}...")
files.download(adapter_zip_filename)

print(f"⬇️ Downloading {processor_zip_filename}...")
files.download(processor_zip_filename)

print("✅ Download initiated for both files.")
```
